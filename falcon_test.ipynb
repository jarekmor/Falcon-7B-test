```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch
from torch import cuda, bfloat16
from accelerate import init_empty_weights
```

```python
device = f'cuda:{cuda.current_device()}' if cuda.is_available() else 'cpu'
```

```python
model = "tiiuae/falcon-7b"
```

```python
print(f"Model loaded on {device}")
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)

pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    return_full_text=True,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=200,
    do_sample=True,
    top_p=0.15,
    top_k=0,
    temperature=0.2,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    repetition_penalty=1.1,
    )
```

```python
sequences = pipeline(
   "Girafatron is obsessed with giraffes, the most glorious animal on the face of this Earth. Giraftron believes all other animals are irrelevant when compared to the glorious majesty of the giraffe.\nDaniel: Hello, Girafatron!\nGirafatron:",
)
```

```python
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```

```python
sequences = pipeline("Explain to me the difference between nuclear fission and fusion.")
```

```python
sequences[0]["generated_text"].split("\n")[1]
```

```python
generate_text = pipeline
```

```python
from langchain import PromptTemplate, LLMChain
from langchain.llms import HuggingFacePipeline
```

```python
# template for an instruction with no input
prompt = PromptTemplate(
    input_variables=["instruction"],
    template="{instruction}"
)
```

```python
llm = HuggingFacePipeline(pipeline=generate_text)
```

```python
llm_chain = LLMChain(llm=llm, prompt=prompt)
```

```python
response = llm_chain.predict(instruction="Explain to me the difference between nuclear fission and fusion.")
```
